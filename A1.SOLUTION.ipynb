# Question : 1

```python
import pandas as pd 
df= pd.read_csv("jsrt_metadata.csv")
df['diagnosis'].value_counts()
```

```python

import matplotlib as plt 
from matplotlib import pyplot 
import numpy as np 
plt.pyplot.hist(df['diagnosis'].value_counts(),bins=50, color='turquoise',edgecolor='black')
plt.pyplot.grid(axis='y', alpha=0.55)
plt.pyplot.title('Histogram of disease diagnosis ')
plt.pyplot.xlabel('Disease Diagnosis')
plt.pyplot.ylabel('Number of patients')
```
